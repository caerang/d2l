# 3.1. Linear Regression

- 회귀는 하나 이상의 독립 변수와 종속 변수 사이의 관계를 모델링하는 방법을 말함
- 자연 과학이나 사회 과학에서는 입력과 출력 사이의 관계의 특성을 결정하기 위해 회귀를 사용하고 기계 학습(Machine Learning)에서는 예측을 하는데 회귀를 사용함
- 회귀는 주택 가격, 주식 가격 예측, 예상 입원일 등 연속적인 값 중에서 예측할 때 발생하는 문제를 다루는데 사용함


## 3.1.1 Basic Elements of Linear Regression

- 선형 회귀는 회귀 방법중 가장 인기가 많고 단순한 방법
- 선형 회귀는 두 가지 전재를 하고 있는데, 첫 번째는 독립 변수와 종속 변수 사이의 관계가 선형성을 갖는 다는 것
- 즉, $y$ 는 $x$의 원소들의 가중치 합으로 표현할 수 있음
- 두 번째는 노이즈가 가우시안 분포를 따른 다고 가정함

- 구체적으로 선형 회귀를 살펴 보기 위해 주택 가격을 예측해야 한다고 가정하자
- 주택의 나이(건축한지 몇 년 되었는지?)와 면적으로 가격을 예측하는 모델을 개발한다고 하자
- 가격을 예측하기 위해서 현재 거래되고 있는 또는 과거에 거래된 주택 정보를 수집해야 한다. 즉, 면적이 얼마이고 몇 년도에 건축된 주택이 얼마에 거래 되었는지에 대한 데이터를 수집해야 함. 모델 개발에 사용할 이러한 데이터를 학습 데이터(training dataset, training set) 이라고 함.
수집한 데이터에서 개별 주택에 대한 데이터(면적, 건축 년도, 가격)를 example(data poing, data instance, sample)라고 부름. 모델이 예측해야 하는 가격은 label(target) 라고 부름
- 모델 개발을 위해 수집한 전체 데이터를 $ x \in X $ 라고 하고, $ x^{(i)} = [x^{(i)}_1, x^{(i)}_2]^{T} $
label 은 $ y^{(i)} $ 로 표기 함

```python
%matplotlib inline
from matplotlib import pyplot as plt
from IPython import display
import torch
import math

x = torch.arange(-7, 7, 0.01)
parameters = [(0, 1), (0, 2), (3, 1)]

display.set_matplotlib_formats('svg')
plt.figure(figsize=(10,6))
for (mu, sigma) in parameters:
    p = (1/math.sqrt(2 * math.pi * sigma**2)) * torch.exp(-(0.5/sigma**2) * (x-mu)**2)
    plt.plot(x.numpy(), p.numpy(), label='mean {}, variance {}'.format(mu, sigma))
plt.legend()
plt.grid()
plt.show()
```
